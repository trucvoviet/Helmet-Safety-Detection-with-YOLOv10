````python
# create yolo model
from ultralytics import YOLO

MODEL_PATH = ('/content/yolov10/yolov10n.pt')
model = YOLO(MODEL_PATH)
````

````python
model.info()
````

````python
YAML_PATH = '/content/safety_helmet_dataset/data.yaml'
EPOCHS = 50
IMG_SIZE = 640
BATCH_SIZE = 64 # if out of memory:1,8,16,32,64

model.train(data=YAML_PATH,
            epochs=EPOCHS,
            batch=BATCH_SIZE,
            imgsz=IMG_SIZE)
````

````python
# evaluate model
TRAINED_MODEL_PATH = 'runs/detect/train2/weights/best.pt'
model = YOLO(TRAINED_MODEL_PATH)

model.val(data=YAML_PATH,
          imgsz=IMG_SIZE,
          split='test')
````

````python
# run prediction with an image
from google.colab.patches import cv2_imshow

TRAINED_MODEL_PATH = 'runs/detect/train2/weights/best.pt'
model = YOLO(TRAINED_MODEL_PATH)

IMAGE_URL = 'https://ips-dc.org/wp-content/uploads/2022/05/Black-Workers-Need-a-Bill-of-Rights.jpeg'
CONF_THRESHOLD = 0.3
results = model.predict(source=IMAGE_URL,
                       imgsz=IMG_SIZE,
                       conf=CONF_THRESHOLD)
annotated_img = results[0].plot()

cv2_imshow(annotated_img)
````
